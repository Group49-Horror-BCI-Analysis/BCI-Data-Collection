```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
andrew_audio = pd.read_csv('../mindwave_data/andrew_audio.csv')
andrew_noaudio = pd.read_csv('../mindwave_data/andrew_noaudio.csv')
kevin_audio = pd.read_csv('../mindwave_data/kevin_audio.csv')
kevin_noaudio = pd.read_csv('../mindwave_data/kevin_noaudio.csv')
richard_audio = pd.read_csv('../mindwave_data/richard_audio.csv')
richard_noaudio = pd.read_csv('../mindwave_data/richard_noaudio.csv')
```

```python
dfs = [andrew_audio, andrew_noaudio, kevin_audio, kevin_noaudio, richard_audio, richard_noaudio]
```

```python
for df in dfs:
    df.rename(columns = {'Unnamed: 0': 'clip'}, inplace = True)
    df.drop(columns = ['raw_value'], inplace = True)
    df.drop_duplicates(inplace = True)
    df['clip'] = df['clip'].str.replace('clip', '')
    df['clip'] = pd.to_numeric(df['clip'].astype(str))
    df.dropna(inplace = True)
    #indexNames = df[(df['attention'] == 0.0) & (df['meditation'] == 0.0)].index
    #df.drop(indexNames, inplace=True)
    #df.drop(df[df['attention'] == 0].index , inplace = True)
    #df.drop(df[df['meditation'] == 0].index , inplace = True)
```

```python
kevin_audio_clips = [kevin_audio[kevin_audio['clip'] == i] for i in range(1,11)]
kevin_noaudio_clips = [kevin_noaudio[kevin_noaudio['clip'] == i] for i in range(1,11)]
kevin = []
for i in range(10):
    df = pd.concat([kevin_audio_clips[i], kevin_noaudio_clips[i]], keys = ['audio', 'no audio'])
    df.index = df.index.droplevel(1)
    kevin.append(df)
```

```python
richard_audio_clips = [richard_audio[richard_audio['clip'] == i] for i in range(1,11)]
richard_noaudio_clips = [richard_noaudio[richard_noaudio['clip'] == i] for i in range(1,11)]
richard = []
for i in range(10):
    df = pd.concat([richard_audio_clips[i], richard_noaudio_clips[i]], keys = ['audio', 'no audio'])
    df.index = df.index.droplevel(1)
    richard.append(df)
```
